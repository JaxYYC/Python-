```python
a=pd.DataFrame({"id":[1001,1002,1003,1004,1005,1006,1007,1008],
                  "gender":['male','female','male','female','male','female','male','female'],
                  "pay":['Y','N','Y','Y','N','Y','N','Y',],
                  "m-point":[10,12,20,40,40,40,30,20]})
b = pd.DataFrame({"id":[1001,1002,1003,1004,1005,1006],
                   "date":pd.date_range('20130102', periods=6),
                   "city":['Beijing ', 'SH', ' guangzhou ', 'Shenzhen', 'shanghai', 'BEIJING '],
                   "age":[23,44,54,32,34,32],
                   "category":['100-A','100-B','110-A','110-C','210-A','130-F'],
                   "price":[1200,np.nan,2133,5433,np.nan,4432]},
                   columns =['id','date','city','category','age','price'])
```

```python
df4 = pd.merge(a,b,how='inner',left_index=True,right_index=True)
df4 = df4.drop(columns=['id_y'])
df4['price'] = df4['price'].fillna(df4['price'].mean())
df4['city'] = df4['city'].replace(['SH'],['Shanghai']).map(str.strip)
df4['city'] = df4['city'].str.lower()
split = pd.DataFrame((x.split('-') for x in df4['category']),index=df4.index,columns=['cat','size'])
df4 = pd.merge(df4,split,how='inner',left_index=True,right_index=True).drop(columns=['category'])
df4
```
